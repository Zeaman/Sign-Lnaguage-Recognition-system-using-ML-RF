```python
import os
import cv2
import numpy as np
import pickle
```

```python
# Set up the video capture
cap = cv2.VideoCapture(0)

# Set the directory to save the images
save_dir = 'sign_language_dataset'

# Set the label for the current gesture
label = input("Enter the label for this gesture: ")

# Create a directory for the label if it doesn't already exist
label_dir = os.path.join(save_dir, label)
if not os.path.exists(label_dir):
    os.makedirs(label_dir)

# Initialize the counter for the number of images captured
count = 0

# Start the video capture loop
while True:
    # Capture a frame from the video stream
    ret, frame = cap.read()

    # Display the frame
    cv2.imshow('Sign Language Recognition', frame)

    # Wait for the user to press a key
    key = cv2.waitKey(1)

    # If the user pressed the 's' key, save the image
    if key == ord('s'):
        # Construct the filename for the image
        filename = os.path.join(label_dir, f'{count}.jpg')
        #if ypu need the label name
        #filename = os.path.join(label_dir, f'{label}{count}.jpg')

        # Convert the image to grayscale
        #gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Save the image to disk (in color)
        cv2.imwrite(filename, frame)

        # Save the image to disk
        #cv2.imwrite(filename, gray)

        # Increment the counter
        count += 1

        # Print a message indicating the image was saved
        print(f"Saved image {filename}")

    # If the user pressed the 'q' key, exit the loop
    elif key == ord('q'):
        break

# Release the video capture
cap.release()

# Close all windows
cv2.destroyAllWindows()
```
